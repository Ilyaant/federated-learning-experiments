```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import pickle
```

```python
def plot_acc_loss(acc, loss, rounds, distribution, sharing=False):
    plt.figure(figsize=(16, 5))

    plt.subplot(1, 2, 1)
    plt.title("Model accuracy")
    plt.xlabel("Communication round")
    plt.ylabel("Accuracy")
    plt.grid()
    plt.plot(rounds, acc)
    
    plt.subplot(1, 2, 2)
    plt.title("Model loss")
    plt.xlabel("Communication round")
    plt.ylabel("Loss")
    plt.grid()
    plt.plot(rounds, loss)
    
    title = f'FL with data distribution: {distribution}'
    if sharing:
        title = f'FL with data distribution: {distribution}\nData sharing implemented'
    plt.suptitle(title)
    plt.show()
```

```python
from torch import nn

class CIFARMediumCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, 3, padding=1), nn.BatchNorm2d(32), nn.LeakyReLU(),
            nn.Conv2d(32, 32, 3, padding=1), nn.BatchNorm2d(32), nn.LeakyReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, 3, padding=1), nn.BatchNorm2d(64), nn.LeakyReLU(),
            nn.Conv2d(64, 64, 3, padding=1), nn.BatchNorm2d(64), nn.LeakyReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64, 128, 3, padding=1), nn.BatchNorm2d(128), nn.LeakyReLU(),
            nn.Conv2d(128, 128, 3, padding=1), nn.BatchNorm2d(128), nn.LeakyReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(128, 256, 3, padding=1), nn.BatchNorm2d(256), nn.LeakyReLU(),
            nn.MaxPool2d(2),
        )
        self.classifier = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(256 * 2 * 2, 512), nn.LeakyReLU(), nn.Dropout(0.5),
            nn.Linear(512, 100)
        )

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)
```

```python
import torch
from torchviz import make_dot

# 1. Define your model
model = CIFARMediumCNN()
dummy_input = torch.randn(1, 3, 32, 32) # Match your input shape

# 2. Pass input through model
output = model(dummy_input)

# 3. Generate and save the figure
dot = make_dot(output, params=dict(model.named_parameters()))
dot.format = 'png'
dot.render('model_architecture')
```
