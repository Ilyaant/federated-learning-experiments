```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
rounds = [i for i in range(1, 9)]
```

```python
def plot_acc_loss(acc, loss, rounds):
    plt.figure(figsize=(16, 5))

    plt.subplot(1, 2, 1)
    plt.title("Динамика accuracy")
    plt.xlabel("Раунд обучения")
    plt.ylabel("Accuracy")
    plt.grid()
    plt.plot(rounds, acc)
    
    plt.subplot(1, 2, 2)
    plt.title("Динамика loss")
    plt.xlabel("Раунд обучения")
    plt.ylabel("Loss")
    plt.grid()
    plt.plot(rounds, loss)
    
    # plt.suptitle(title)
    plt.show()
```
